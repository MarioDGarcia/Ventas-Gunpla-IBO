## **Limpiando ID's (primera parte)**


```python
import pandas as pd
```

```python
df = pd.read_csv("Ventas_gunplas_sucias.csv")
df.head()
```

```python
df.tail()
```

```python
df["Id_Venta"] = df.index
df.head()
```

```python
del df["id_venta"]
df.head()
```

### ***Limpiando id de prodcutos***

```python
productos = pd.read_csv("IBO_Gunplas.csv")
productos.head()
```

```python
productos.describe()
```

```python
df = df[df["id_producto"] <= 65] ## se puede realizar un join/merge entre ambos archivos y
                                # que se haga autoamtico, pero este es otro metodo
```

```python
df[df["id_producto"]>64] ## comprobamos si esta el ultimo producto (el 65)
```

## **Cambiar a int el id Prodcuto**

```python
df["id_producto"] = df["id_producto"].astype(int)
df.head()
```

## **Eliminar Columnas NaN**

```python
del df["Id_Pais"]
df
```

```python
del df["Fecha_venta"]
df
```

## **ID PAIS**

```python
paises = pd.read_csv("Paises con continene.csv")
paises
```

```python
listaPaisesCodigo = paises["id_pais"].unique().tolist() ## otra forma seria el join/merge
listaPaisesCodigo
```

```python
df = df[df["id_pais"].isin(listaPaisesCodigo)]
```

```python
df
```

## **Unidades**
En este apartado existen cadenas de texto como "muchas" en lugar de un numero, asi que vamos a convertir todo a numero

```python

df['unidades_limpio'] = df['unidades'].str.extract('(\d+)') 
# con expresion regular extraemos solo el numero si es que hay mezcla (2 unidades, 22 piezas, etc)


df['unidades_limpio'] = pd.to_numeric(df['unidades_limpio'], errors='coerce')

```

## **Eliminar Nulos**

```python
df = df.dropna()

```

```python
df[df.isna().any(axis=1)]
```

```python
del df["unidades"]
```

```python
df["unidades_limpio"] = df["unidades_limpio"].astype(int)
```

## **Dar Id de Venta**

```python
del df["Id_Venta"] 
```

```python
df['Id_venta'] = range(1, len(df) + 1)
df
```

```python
df.index = df["Id_venta"]
df
```

```python
df['fecha_venta'] = pd.to_datetime(df['fecha_venta'], errors='coerce', dayfirst=False)
```

```python
del df["Id_venta"]
```

```python
df.rename(columns={'unidades_limpio': 'unidades'}, inplace=True)
```

```python
df.to_csv("Ventas_IBO_Gunpla_V2.0.csv") ## V2.0 porque es la segunda vez que se realizaba esta peque√±a practica
```
